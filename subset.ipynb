        - df.head(), df.tail(),df.sample()

```python
import pandas as pd
import numpy as np

# df = pd.DataFrame({
#     "a":[4,5,6,6,np.nan],
#     "b":[7,8,np.nan,9,9],
#     "c":[10,11,12,np.nan,12],
# }, index=[1,2,3,4,5])
df = pd.DataFrame({
    "a":[4,5,6,6,np.nan],
    "b":[7,8,np.nan,9,9],
    "c":[10,11,12,np.nan,12],
}, index=list(range(1,6)))

df
```

```python
range(1,6)
list(range(1,6))
```

```python
# 일부분 행 추출 - df.head(n)/ df.tail(n)
df.head()
```

```python
# 임의로 하나 추출
print(df.sample())
print(df.sample(2))
print(df.sample(frac=0.5))
```

```python
# b 컬럼에서 가장 큰 거 한 행

df.nlargest(1,"b")
```

```python
df.nsmallest(2,"a")
```

```python
import seaborn as sns

df = sns.load_dataset("iris")
df.head(3)
```

```python
# 컬럼 추출

df["petal_length"]
```

```python
df[["petal_length","sepal_width"]]
```

```python
columns=["petal_length","sepal_width","species"]
df[columns].tail(3)
```

```python
# 행,열
# loc[], iloc[]
df.columns

# df[['sepal_length','sepal_width','petal_length']]

# df.loc[행,열] : 행열 표현시 :

df.loc[:,'sepal_length':'petal_length']

# 2~5행, 컬럼 추출
df.loc[2:5,'sepal_length':'petal_length']
```

```python
# index : 인덱스 번호 이용
# 뒤에 인덱스값은 미포함

# 행은 10~19
```

```python
df[df["petal_length"]>3]
```

```python
# 행: sepal_length  >3 큰 행만 추출하고, 컬럼은 sepla_length,sepal_sidth 추출

df.loc[df["sepal_length"]] >5 , ["sepal_length","sepal_width"]
```

```python
friend_dict_list =[
    {
        "name":"John",
        "age":25,
        "job":"student"
    },
    {
        "name":"Nate",
        "age":30,
        "job":"teacher"
    },
    {
        "name":"Jenny",
        "age":30,
        "job":"developer"
    }
]
friend_df = pd.DataFrame(friend_dict_list)
friend_df
```

```python
# Nate,Jenny 행만 추출
friend_df.loc[1:2]
friend_df.iloc[1:3]
```

```python
# John, Jenny 행 , name,job 컬럼 추출
friend_df.loc[[0,2],["name","job"]]
friend_df.iloc[[0,2],[0,2]]
```

```python
# 이름이 Nate 이고 나이가 25보다 많은 행 추출

friend_df[(friend_df.name == "Nate") & (friend_df.age >25)]
```

```python
friend_df.query("age>25")
```

```python
# python : and,or,not
# pandas : & , |, !
friend_df.query("age > 25"and "name == 'Nate'")
```
